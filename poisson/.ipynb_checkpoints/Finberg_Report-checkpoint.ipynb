## Convergence Comparison

<span style='font-family:"Times New Roman"'> <span styel=''>
From figures 9 and 10, it can be seen that the Gauss-Seidel method converges the slowest in time while both the Gauss-Seidel and the Jacobi methods convergence comparably slow in terms of iterations. Conversely, the Conjugate Gradient solver is the fastest to converge in terms of iterations, iterating only once for these relatively coarse grids. Additionally, all solvers except Gauss-Seidel converged comparably quickly in time. While the direct solver converged relatively quickly in time in these test cases, this convergence time would not hold if the grid were to be refined more.

<span style='font-family:"Times New Roman"'> <span styel=''>
Key:
    
    Star = Dirichlet
    
    Triangle = Neumann

```python
solvers = [ 'Jacobi', 'Gauss-Seidel', 'Conjugate Gradient', 'Direct Solver' ]
D = [ 6979, 5110, 1, 1 ]
N = [ 6999, 6856, 1, 1]

for i in range( len( D ) ):
    pyplot.plot( solvers[ i ], D[ i ], '*', markersize = 10 )
    pyplot.plot( solvers[ i ], N[ i ], '^', markersize = 10 )
    
pyplot.ylabel( 'Iterations' )
pyplot.grid()
pyplot.xlabel( 'Solvers' )
pyplot.title( '100x100 Grid' );
```

                       Figure 9

```python
D = [ 1.0393, 185.79, 2.9998, 2.9132 ]
N = [ 1.0712, 254.0178, 2.9567, 3.0379 ]

for i in range( len( D ) ):
    pyplot.plot( solvers[ i ], D[ i ], '*', markersize = 10 )
    pyplot.plot( solvers[ i ], N[ i ], '^', markersize = 10 )
    
pyplot.ylabel( 'Time (s)' )
pyplot.grid()
pyplot.xlabel( 'Solvers' )
pyplot.title( '100x100 Grid' );
```
